# Transfer Learning and Fine-Tuning

- Learns how to classify images of cats and dogs by using transfer learning from a pre-trained network
- Practice two methods to customize a pre-trained model which are **feature extraction** and **fine-tuning**
- Full tutorial reference: https://www.tensorflow.org/tutorials/images/transfer_learning

````python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

print('TensorFlow version:', tf.__version__)
````

## Data Preprocessing
A dataset contains several thousand images of cats and dogs. As same as before, use ```tf.keras.utils.image_dataset_from_directory``` to create both train and validation ```tf.data.Dataset```.

````python
dataset_url = 'https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip'
path_to_zip = tf.keras.utils.get_file('cats_and_dogs.zip', origin=dataset_url, extract=True)
PATH = os.path.join(os.path.dirname(path_to_zip), 'cats_and_dogs_filtered')
````

````python
train_dir = os.path.join(PATH, 'train')
val_dir = os.path.join(PATH, 'validation')
````

````python
batch_size = 32
img_size = (160, 160)

train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    batch_size=batch_size,
    image_size=img_size
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    val_dir,
    batch_size=batch_size,
    image_size=img_size
)
````

Use ```Dataset.class_names``` to check how many classes are in inside a dataset. Here, we only expect two classes, cats and dogs.

````python
class_names = train_dataset.class_names
print(class_names)
````

````python
plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
    for i in range(9):
        plt.subplot(3, 3, i+1)
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.title(class_names[labels[i]])
        plt.axis('off')

plt.show()
````

One you might notice that there is no test images in this dataset. Therefore, use ```tf.data.experimental.cardinality``` to see how many batches of data are available in a validation dataset (hint: we defined batch_size=32 for val_dataset). Then, make 20% of them to a test dataset.

````python
val_batches = tf.data.experimental.cardinality(val_dataset)
print(val_batches.numpy())
````

````python
test_dataset = val_dataset.take(val_batches // 5)
val_dataset = val_dataset.skip(val_batches // 5)
print('Number of validation and test batches: {}, {}.'.format(
    tf.data.experimental.cardinality(val_dataset).numpy(), tf.data.experimental.cardinality(test_dataset).numpy()))
````

Configure the dataset for performance as what we always do throughout the tutorials. Also, use **data augmentation techique** to create sample diversity by applying random, yet identifiable transformations to the training images. Data augmentation is powerful method when there is a few images in the dataset to train the model.

````python
AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
````

````python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2)
])
````

````python
plt.figure(figsize=(10, 10))
for images, _ in train_dataset.take(1):
    for i in range(9):
        plt.subplot(3, 3, i+1)
        augmented_images = data_augmentation(images)
        plt.imshow(augmented_images[0].numpy().astype('uint8'))
        plt.axis('off')

plt.show()
````

## Create the Base Model
The base model, **MobileNet V2** developed at Google, is a pre-trained model on the ImageNet dataset, a large dataset consisting of 1.4M images and 1000 classes. Since the base model expects to have an input pixel values in range of [-1, 1], we need to rescale pixel values from [0, 255] to [-1, 1] using ```tf.keras.applications.mobilenet_v2.preprocess_input```.

````python
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
````

````python
img_shape = img_size + (3,)
base_model = tf.keras.applications.MobileNetV2(
    input_shape=img_shape,
    include_top=False, # if False, the loaded model does not have the classification layers at the top
    weights='imagenet' # trained on the ImageNet dataset
)
````

This feature extractor (base_model) converts each (160, 160, 3) image to a (5, 5, 1280) block of features.

````python
image_batch, label_batch = next(iter(train_dataset))
feature_batch = base_model(image_batch)
print(feature_batch.shape)
````

## Feature extraction
Here, we will freeze the convolutional base and use as a **feature extractor**. To do so, add a classifier on top of the feature extractor and train the top-level classifier.

````python
base_model.trainable = False
````

````python
base_model.summary()
````

## Add a Classification Head

The output shape from the base model is (batch_size, 5, 5, 1280). There are 5x5 = 25 feature maps, and in order to predict the class labels, ```tf.keras.layers.GlobalAveragePooling2D``` layer is required to average all 25 feature maps into a single 1280-element vector per image.

````python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
````

After that, apply a ```tf.keras.layers.Dense``` layer to convert 1280 single-vector (1-D) features into a single prediction per image.

````python
prediction_layer = tf.keras.layers.Dense(1)
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
````

Build a complete model by chaining ```data_augmentation```, ```preprocess_input```, ```base_model```, ```global_average``` and ```prediction``` layers (also a dropout layer) together using the Keras functional API.

````python
inputs = tf.keras.Input(shape=(160, 160, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
````

## Compile and Train the Model

````python
base_learning_rate = 0.0001
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), # since there are binary classes, use BinaryCrossentropy
    metrics=['accuracy']
)
````

````python
model.summary()
````

````python
epochs = 10

initial_eval = model.evaluate(val_dataset)
````

````python
print('initial loss and accuracy: {:.2f}, {:.2f}'.format(initial_eval[0], initial_eval[1]))
````

````python
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs
)
````

## Visualize the Training Results
Display learning curves of the training and validation accuracy/loss when using the MobileNetV2 base model as a fixed feature extractor.

````python
loss = history.history['loss']
val_loss = history.history['val_loss']

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

num_of_epochs = history.epoch

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.subplots_adjust(hspace=0.3)
plt.plot(num_of_epochs, loss, label='train loss')
plt.plot(num_of_epochs, val_loss, label='val loss')
plt.grid(True)
plt.xticks(num_of_epochs)
plt.ylim([0, 1])
plt.legend(loc='upper right')
plt.title('train & validation loss vs. # of epochs')

plt.subplot(2, 1, 2)
plt.plot(num_of_epochs, acc, label='train acc')
plt.plot(num_of_epochs, val_acc, label='val acc')
plt.grid(True)
plt.xticks(num_of_epochs)
plt.ylim([min(plt.ylim()), 1])
plt.legend(loc='lower right')
plt.title('train & validation accuracy vs. epochs')

plt.show()
````

## Fine-Tuning
In the feature extraction part, only few layers on top of the base model were trained, and the weights of the pre-trained network MobileNet V2 were **not updated** during training. Fine-tuning comes from the idea that not only training the classifier on top of the model, but also fine-tuning the weights of the other layers of the pre-trained network to enhance model's performance even better.

## Unfreeze the Top Layers of the Model
All we need to do is unfreeze the ```base_model``` and set the bottom layers to be a **non-trainable**.

````python
base_model.trainable = True
````

````python
print('# of layers in the base model:', len(base_model.layers))

fine_tune_at = 100
for layer in base_model.layers[0:fine_tune_at]:
    layer.trainable = False
````

## Re-compile and Re-train the Model
It is critical to use **very low learning rate**, because you are training a much larger model. If the learning rate (hint: learning rate is the hyper-parameter which set the magnitude of how much the weights will update) is large, the model is at risk of overfitting.

````python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate/10),
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), # same as before
    metrics=['accuracy']
)
````

````python
model.summary()
````

````python
fine_tune_epochs = 10
total_epochs = epochs + fine_tune_epochs

history_fine_tuned = model.fit(
    train_dataset,
    epochs=total_epochs,
    initial_epoch=history.epoch[-1],
    validation_data=val_dataset
)
````

## Visualize the Training Results after Fine-Tuning

````python
loss = history.history['loss'] + history_fine_tuned.history['loss']
val_loss = history.history['val_loss'] + history_fine_tuned.history['val_loss']

acc = history.history['accuracy'] + history_fine_tuned.history['accuracy']
val_acc = history.history['val_accuracy'] + history_fine_tuned.history['val_accuracy']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.subplots_adjust(hspace=0.3)
plt.plot(loss, label='train loss')
plt.plot(val_loss, label='val loss')
plt.axvline(x=9, color='r', ls='--', label='Start Fine Tuning')
plt.grid(True)
plt.xticks(num_of_epochs)
plt.ylim([0, 1])
plt.legend(loc='upper right')
plt.title('train & validation loss vs. # of epochs')

plt.subplot(2, 1, 2)
plt.plot(acc, label='train acc')
plt.plot(val_acc, label='val acc')
plt.axvline(x=9, color='r', ls='--', label='Start Fine Tuning')
plt.grid(True)
plt.xticks(num_of_epochs)
plt.ylim([min(plt.ylim()), 1])
plt.legend(loc='lower right')
plt.title('train & validation accuracy vs. epochs')

plt.show()
````

## Evaluation and Prediction
In order to verify the performance of the model on new data, use a test dataset that we created before.

````python
test_eval = model.evaluate(test_dataset)
print('Test loss and accuracy: {:.2f}, {:.2f}'.format(test_eval[0], test_eval[1]))
````

```flatten()``` returns a numpy array of prediction that collapsed into one dimension. After ```one_batch_prediction``` is flatten, apply a ```tf.nn.sigmoid``` function since the model returns logits. ```tf.nn.sigmoid``` is used to decide between two classes.

````python
image_batch, label_batch = test_dataset.as_numpy_iterator().next()
one_batch_prediction = model.predict_on_batch(image_batch).flatten()

one_batch_prediction = tf.nn.sigmoid(one_batch_prediction)
one_batch_prediction = tf.where(one_batch_prediction < 0.5, 0, 1)

plt.figure(figsize=(10, 10))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.imshow(image_batch[i].astype('uint8'))
    plt.title(class_names[one_batch_prediction[i]])
    plt.axis('off')
    
plt.show()
````

## Summary
- **Using a pre-trained model for feature extraction**: It is a common practice to take advantage of features learned by a model trained on a larger dataset compared to what we have. The convolutional base model extracts features associated with each images and all we have to do is training the new classifier on top of the base model which determines the image classes (whether the image belongs to cats or dogs).

- **Fine-tuning a pre-trained network**: To improve better output results, fine-tuning the top-level layers of a pre-trained network and classifier together is another option you can take into account. In this case, we tuned the weights such that the model learned high-level features specific to the new dataset.

To learn more about fine-tuning, visit [TensorFlow guide on transfer learning](https://www.tensorflow.org/guide/keras/transfer_learning).
