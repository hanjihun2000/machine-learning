# Stable Diffusion using KerasCV's Model
- Shows how to generate images using the KerasCV implementation of [Stable Diffusion](https://github.com/CompVis/stable-diffusion). 
- Full tutorial reference: https://www.tensorflow.org/tutorials/generative/generate_images_with_stable_diffusion 

Stable Diffusion is a powerful, open-source text-to-image generation model to produce novel and synthetic images based on a text prompt you typed.

````python
from tensorflow import keras
import keras_cv
import matplotlib.pyplot as plt
````

Construct a model by calling ```keras_cv.models.StableDiffusion()``` with image settings.

````python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
````

````python
def plot_images(images):
  plt.figure(figsize=(20, 20))
  for i in range(len(images)):
    plt.subplot(1, len(images), i+1)
    plt.imshow(images[i])
    plt.axis("off")
````

Use ```model.text_to_image()``` to give the model a text prompt and the size of images to generate.

````python
diffusion_images = model.text_to_image("a photograph of a girl wearing Korean traditional costume called hanbok", batch_size=3)
````

````python
plot_images(diffusion_images)
````
