# CIFAR-10 CNN Model

- CIFAR-10 Dataset: https://www.cs.toronto.edu/~kriz/cifar.html

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.callbacks import TensorBoard
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
cifar10 = tfds.load('cifar10', with_info=False, shuffle_files=False)
```

```python
class_labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
cifar10_train = cifar10['train']
cifar10_test = cifar10['test']
```

```python
print(len(cifar10_train))
```

```python
print(len(cifar10_test))
```

```python
cifar_val = cifar10_train.take(10000)
cifar10_train = cifar10_train.skip(10000)
```

```python
# data augmentation
tf.random.set_seed(1)
BUFFER_SIZE = 40000
BATCH_SIZE = 64
NUM_EPOCHS = 50
STEPS_PER_EPOCH = np.ceil(len(cifar10_train)/BATCH_SIZE)

def preprocess(sample, mode='train'):
  image = sample['image']
  label = sample['label']
  if mode == 'train':
    image_flip_up_down = tf.image.random_flip_up_down(image)
    image_flip_all = tf.image.random_flip_left_right(image_flip_up_down)
    return tf.cast(image_flip_all, tf.float32)/255.0, label
  else:
    return tf.cast(image, tf.float32)/255.0, label

train_ds = cifar10_train.map(lambda x: preprocess(x, mode='train'))
train_ds = train_ds.shuffle(buffer_size=BUFFER_SIZE).repeat().batch(BATCH_SIZE)
val_ds = cifar_val.map(lambda x: preprocess(x, mode='val')).batch(BATCH_SIZE)
```

```python
# display cifar10 images
batch = train_ds.take(1)
```

```python
fig = plt.figure(figsize=(16, 8))
for element in batch.as_numpy_iterator():
  for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(element[0][i])
    ax.set_title('{}'.format(class_labels[element[1][i]]), size=15)
plt.tight_layout()
plt.show()
```

```python
# create model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    
    tf.keras.layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
])
```

```python
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(10))
model.summary()
```

```python
callbacks = [
    ModelCheckpoint(filepath='cifar10-earlystopping.h5', monitor='val_loss'),
    EarlyStopping(patience=3, restore_best_weights=True),
    TensorBoard()
]

model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
history = model.fit(train_ds, epochs=20, validation_data=val_ds, steps_per_epoch=STEPS_PER_EPOCH)
```

```python
test_ds = cifar10_test.map(lambda x: preprocess(x, mode='test')).batch(BATCH_SIZE)
result = model.evaluate(test_ds)
print('eval loss: {}, eval acc: {}'.format(*result))
```

```python
num_epochs = len(history.history['accuracy'])
```

```python
# more training
history = model.fit(train_ds, epochs=num_epochs+50, validation_data=val_ds, initial_epoch=num_epochs, steps_per_epoch=STEPS_PER_EPOCH, callbacks=callbacks)
```

```python
# display 10 cifar10 test images with actual label and predicted value
samples = test_ds.unbatch().take(10)
```

```python
logits = model.predict(samples.batch(1))
preds = tf.argmax(logits, axis=1)
print(preds)
```

```python
fig = plt.figure(figsize=(16, 8))
for element in samples.enumerate():
  ax = fig.add_subplot(2, 5, element[0].numpy()+1)
  ax.set_xticks([])
  ax.set_yticks([])
  # ax.set_title('{}'.format(class_labels[element[1][1].numpy()], size=15, color='blue'))
  ax.set_title('Real: {}, Pred: {}'.format(class_labels[element[1][1].numpy()], class_labels[preds[element[0].numpy()]]), size=15, color='blue')
  ax.imshow(element[1][0].numpy())
  # ax.text(0.9, 0.1, '{}'.format(class_labels[preds[element[0].numpy()]]), size=15, color='red', horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
plt.tight_layout()
plt.show()
```
